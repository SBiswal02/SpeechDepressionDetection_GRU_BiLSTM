```python
import os
import shutil
```

```python
def copy_audios(source_directory, destination_folder):
    # Ensure the destination folder exists
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate over all items in the source directory
    for item in os.listdir(source_directory):
        # Construct the full path of the item
        item_path = os.path.join(source_directory, item)

        # Check if the item is a directory and matches the pattern "num_P"
        if os.path.isdir(item_path) and item.endswith("_P"):
            # Construct the expected audio file name (e.g., "1_AUDIO.wav")
            audio_file_name = item.replace("_P", "_AUDIO.wav")
            audio_file_path = os.path.join(item_path, audio_file_name)

            # Check if the audio file exists
            if os.path.isfile(audio_file_path):
                # Copy the audio file to the destination folder
                shutil.copy(audio_file_path, destination_folder)
                print(f"Copied {audio_file_name} to {destination_folder}")
            else:
                print(f"Audio file {audio_file_name} not found in {item_path}")
```

```python
def copy_transcripts(source_directory, destination_folder):
    # Ensure the destination folder exists
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate over all items in the source directory
    for item in os.listdir(source_directory):
        # Construct the full path of the item
        item_path = os.path.join(source_directory, item)

        # Check if the item is a directory and matches the pattern "num_P"
        if os.path.isdir(item_path) and item.endswith("_P"):
            # Construct the expected transcript file name (e.g., "1_TRANSCRIPT.txt")
            transcript_file_name = item.replace("_P", "_Transcript.csv")
            transcript_file_path = os.path.join(item_path, transcript_file_name)

            # Check if the transcript file exists
            if os.path.isfile(transcript_file_path):
                # Copy the transcript file to the destination folder
                shutil.copy(transcript_file_path, destination_folder)
                print(f"Copied {transcript_file_name} to {destination_folder}")
            else:
                print(f"Transcript file {transcript_file_name} not found in {item_path}")
```

```python
def copy_transcripts_allcaps(source_directory, destination_folder):
    # Ensure the destination folder exists
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate over all items in the source directory
    for item in os.listdir(source_directory):
        # Construct the full path of the item
        item_path = os.path.join(source_directory, item)

        # Check if the item is a directory and matches the pattern "num_P"
        if os.path.isdir(item_path) and item.endswith("_P"):
            # Construct the expected transcript file name (e.g., "1_TRANSCRIPT.txt")
            transcript_file_name = item.replace("_P", "_TRANSCRIPT.csv")
            transcript_file_path = os.path.join(item_path, transcript_file_name)

            # Check if the transcript file exists
            if os.path.isfile(transcript_file_path):
                # Copy the transcript file to the destination folder
                shutil.copy(transcript_file_path, destination_folder)
                print(f"Copied {transcript_file_name} to {destination_folder}")
            else:
                print(f"Transcript file {transcript_file_name} not found in {item_path}")
```

```python
# Example usage
destination_folder = "/home/smrutibiswal/Research/Dataset/data/train/audios"
source_directory = "/home/smrutibiswal/Research/Dataset/data/train/dataset"

copy_audios(source_directory, destination_folder)
```

```python
destination_folder_transcripts = "/home/smrutibiswal/Research/Dataset/data/train/transcripts"
copy_transcripts(source_directory, destination_folder_transcripts)
copy_transcripts_allcaps(source_directory, destination_folder_transcripts)
```

```python
# rename all Transcripts to TRANSCRIPT.csv
for filename in os.listdir(destination_folder_transcripts):
    if filename.endswith(".csv"):
        os.rename(os.path.join(destination_folder_transcripts, filename), os.path.join(destination_folder_transcripts, filename.replace("Transcript", "TRANSCRIPT")))
        print(f"Renamed {filename} to {filename.replace('Transcript', 'TRANSCRIPT')}")
    else:
        continue
    
```

```python
validation_source_directory = "/home/smrutibiswal/Research/Dataset/data/validation/data"
validation_destination_folder = "/home/smrutibiswal/Research/Dataset/data/validation/audios"

if not os.path.exists(validation_destination_folder):
    os.makedirs(validation_destination_folder)
    
copy_audios(validation_source_directory, validation_destination_folder)

validation_destination_folder_transcripts = "/home/smrutibiswal/Research/Dataset/data/validation/transcripts"
if not os.path.exists(validation_destination_folder_transcripts):
    os.makedirs(validation_destination_folder_transcripts)
    
copy_transcripts(validation_source_directory, validation_destination_folder_transcripts)
copy_transcripts_allcaps(validation_source_directory, validation_destination_folder_transcripts)
```

```python
# rename all Transcripts to TRANSCRIPT.csv

for filename in os.listdir(validation_destination_folder_transcripts):
    if filename.endswith(".csv"):
        os.rename(os.path.join(validation_destination_folder_transcripts, filename), os.path.join(validation_destination_folder_transcripts, filename.replace("Transcript", "TRANSCRIPT")))
        print(f"Renamed {filename} to {filename.replace('Transcript', 'TRANSCRIPT')}")
```

```python
test_source_directory = "/home/smrutibiswal/Research/Dataset/data/test/data"
test_destination_folder = "/home/smrutibiswal/Research/Dataset/data/test/audios"

if not os.path.exists(test_destination_folder):
    os.makedirs(test_destination_folder)
    
copy_audios(test_source_directory, test_destination_folder)

test_destination_folder_transcripts = "/home/smrutibiswal/Research/Dataset/data/test/transcripts"
if not os.path.exists(test_destination_folder_transcripts):
    os.makedirs(test_destination_folder_transcripts)
    
copy_transcripts(test_source_directory, test_destination_folder_transcripts)
copy_transcripts_allcaps(test_source_directory, test_destination_folder_transcripts)
```
